#### In order to take "timestamp" out of string_list_data, a new variable called "timestamps" is made using the "x for x in" syntax.

```python
timestamps = [x[0]['timestamp'] for x in test2['string_list_data']]
```

```python
timestamps
```

#### I believe "test2['timestamp'] = timestamps" is used so that the timestamps can have its own column. Then a new variable called "final_df" is made using test2.drop(). The final steps taken is to use groupby(), count(), and sort_values() to make the final data frame.

```python
test2['timestamp'] = timestamps
```

```python
final_df = test2.drop('string_list_data', axis = 1)
```

```python
final_df.head()
```

#### Using groupby(), count(), and sort_values(), I made some final data frames. The first data frame displays different Instagram users under "title" along with the amount of timestamps for each user. It looks like the numbers under "timestamp" and "media_list_data" represent the amount of times I've like each user's posts, and that type of data could be reliable if Instagram wants to know the type of content I tend to view the most. They might factor in which accounts I liked the most when deciding which posts should be displayed in my recommended feed. One potential problem is that this data frame doesn't show the exact timestamps for each post so this could be unreliable if Instagram wants to know which times I tend to use the app.

```python
final_df.groupby('title').count().sort_values('timestamp', ascending=False)
```

#### I decided to see what would happen if I switched 'title' and 'timestamp' and it outputted the exact timestamps for each post.
